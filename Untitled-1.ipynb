# Part1

```python
# Initialize a KMeans clustering model with the specified number of clusters (k)
k = 5
kmeans = KMeans(n_clusters=k)
kmeans.fit(data[['market_cap', 'volume']])
```

```python
#Retrieve the cluster labels assigned to each data point by the KMeans model.
labels = kmeans.labels_
data['cluster'] = labels
```

```python
# Colors for clusters
colors = ['r', 'g', 'b', 'c', 'm']

# Plot points corresponding to each cluster
for i in range(5):
    cluster_data = data[data['cluster'] == i]
    plt.scatter(cluster_data['market_cap'], cluster_data['volume'], label=f'Cluster {i}', c=colors[i])

# Plot cluster centers
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], marker='x', s=100, c='k', label='Cluster Centers')

# Label the axes
plt.xlabel('market_cap')
plt.ylabel('Volume')

# Display cluster labels on the plot
for i in range(5):
    plt.annotate(f'Cluster {i}', (centers[i, 0], centers[i, 1]), textcoords="offset points", xytext=(0, 10), ha='center')

# Show legends
plt.legend()
plt.title('K-Means Clustering')

# Display the plot
plt.show()
```

In the first part of the analysis, we applied the K-Means clustering algorithm using only two features, cap market, and volume, to cluster the dataset into 5 clusters. We then plotted the results on a scatter plot, highlighting which data points belong to each cluster. The center of each cluster is also marked on the plot.

### Elbow Method


```python
k_values = range(1, 11)
elbow_point = []

for k in k_values:
    kmeans_elbow = KMeans(n_clusters=k)
    kmeans_elbow.fit(data[['market_cap', 'volume']])
    wcss = kmeans_elbow.inertia_   # Calculate the Within-Cluster Sum of Squares (WCSS)
    elbow_point.append(wcss)       # Append the WCSS value to elbow_point list for later analysis


plt.figure(figsize=(8, 4))
plt.plot(k_values, elbow_point, marker='o', linestyle='-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.title('Elbow Method for Optimal k')
plt.grid()
plt.show()
```

### Silhouette Method

```python
sil = [] # Create an empty list to store silhouette scores for different k values.
k_values = range(2, 11)


for k in k_values:
  kmeans_sil = KMeans(n_clusters = k).fit(data[['market_cap', 'volume']])
  labels_sil = kmeans_sil.labels_
  sil.append(silhouette_score(data[['market_cap', 'volume']], labels_sil, metric = 'euclidean')) # Calculate the silhouette score for the current k using the 'euclidean' distance metric.
```

```python
# Create a plot to visualize the Silhouette Method results.
plt.figure(figsize=(8, 4))
plt.plot(k_values, sil, marker='o', linestyle='-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('silhouette_score')
plt.title('Silhouette Method for Optimal k')
plt.grid()
plt.show()
```

The optimal k value of 4 was chosen and the modeling is done with 4

```python
#Finding the number of neighbors according to the number of noises and clusters
n_neighbors=range(10, 20)
# Iterate through each value of n_neighbors.
for n in n_neighbors:
    nn=NearestNeighbors(n_neighbors=n)   # Create a Nearest Neighbors (NN) model with the current number of neighbors (n).
    neigh_dist,neigh_ind=nn.fit(data[["market_cap","volume"]]).kneighbors()
    sorted_neigh=np.sort(neigh_dist[:,n-1]) # Sort the distances to the nth nearest neighbor.
    kl=KneeLocator(np.arange(len(sorted_neigh)),sorted_neigh,curve="convex",direction="increasing")  # Use the KneeLocator to find the elbow point in the sorted distances.
    dbscan=DBSCAN(eps=sorted_neigh[kl.elbow],min_samples=n)
    dbscan.fit(data[["market_cap","volume"]])
    print("n_neighbors:{0}   noises:{1}   clusters:{2}".format(n,np.unique(dbscan.labels_,return_counts=True)[1][0],len(np.unique(dbscan.labels_,return_counts=True)[0])-1))
```
