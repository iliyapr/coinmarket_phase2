```python
data = yf.download("XMR-USD", start="2017-11-09", end="2023-09-07")

sma_index = data["Close"].rolling(window=20).mean()
sma_index = pd.DataFrame(sma_index)
sma_index.rename(columns={"Close" : "SMA Index"}, inplace=True)

ema_index = data["Close"].ewm(span=20, adjust=False).mean()
ema_index = pd.DataFrame(ema_index)
ema_index.rename(columns={"Close" : "EMA Index"}, inplace=True)

delta = data["Close"].diff()
gain = delta.where(delta > 0, 0)
loss = -delta.where(delta < 0, 0)
avg_gain = gain.rolling(window=14).mean()
avg_loss = loss.rolling(window=14).mean()
rs = avg_gain / avg_loss
rsi_index = 100 - (100 / (1 + rs))
rsi_index = pd.DataFrame(rsi_index)
rsi_index.rename(columns={"Close" : "RSI Index"}, inplace=True)

rsi_index
```
