# KNN

```python
error_rate = list()
for k in range(1,500):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(X_train , y_train)
    pred_ith = knn_model.predict(X_val)
    error_rate.append(np.mean(pred_ith != y_val))
```

```python
fig = plt.figure(figsize=(15,12))
plt.plot(range(1,500) , error_rate )
```

```python
k = error_rate.index(min(error_rate))
knn_model = KNeighborsClassifier(n_neighbors=k)
knn_model.fit(X_train , y_train)
```

```python
k
```

```python
pred = knn_model.predict(X_val)
```
